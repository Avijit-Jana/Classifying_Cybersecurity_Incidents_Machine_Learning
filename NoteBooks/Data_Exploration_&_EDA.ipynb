# Importing Libraries & load data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Dataset/GUIDE_Train.csv')  # Load the dataset
df.head(3)
```

```python
df.info()   # dataset information
```

```python
# Print the shape of the dataset
print(f"Dataset Shape: {df.shape}\n")
```

# EDA

```python
df.describe()   # summary statistics
```

```python
# Distribution of the target variable
df['IncidentGrade'].value_counts(normalize=True)
```

```python
# Plot the distribution of the target variable
plt.figure(figsize=(10, 4))
sns.countplot(df['IncidentGrade'])  # Replace 'target_column' with the actual name of the target column
plt.title("Distribution of Target Classes (TP, BP, FP)")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
# Select only numerical columns
numerical_data = df.select_dtypes(include=[np.number])

# Calculate correlation matrix for numerical features
correlation_matrix = numerical_data.corr()

plt.figure(figsize=(20, 10))    # Set the size of the plot
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f") # Plot the correlation matrix
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

```python
plt.figure(figsize=(8, 5))      # Set the figure size
sns.countplot(x=df['IncidentGrade'], data=df)                   # Plot the countplot
plt.title(f"Class Distribution of {df['IncidentGrade']}")       # Set the title of the plot
plt.xlabel('IncidentGrade')     # Set the x-label
plt.ylabel("Count")             # Set the y-label
plt.show()
```
